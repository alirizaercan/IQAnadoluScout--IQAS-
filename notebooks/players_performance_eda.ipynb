## Function to remove percentage values in parentheses and convert to float

```python
def remove_percentage(value):
    if isinstance(value, str) and '(' in value:
        return float(value.split('(')[0].strip())
    return value
```

## Apply the function to all columns

```python
for col in df.columns:
    df[col] = df[col].apply(remove_percentage)
```

## Convert numerical columns to appropriate data types

```python
numerical_columns = [
'total_played', 'started', 'minutes_per_game', 'total_minutes_played', 'team_of_the_week', 'goals_conceded_per_game', 'penalties_saved', 'saves_per_game', 'succ_runs_out_per_game', 'goals_conceded', 'conceded_from_inside_box', 'conceded_from_outside_box', 'saves_made', 'goals_prevented', 'saves_from_inside_box', 'saves_from_outside_box', 'saves_caught', 'saves_parried', 'goals', 'scoring_frequency', 'goals_per_game', 'shots_per_game', 'shots_on_target_per_game', 'big_chances_missed', 'goal_conversion', 'penalty_goals', 'penalty_conversion', 'free_kick_goals', 'free_kick_conversion', 'goals_from_inside_the_box', 'goals_from_outside_the_box', 'headed_goals', 'left_foot_goals', 'right_foot_goals', 'penalty_won', 'assists', 'expected_assists_(xa)', 'touches_per_game', 'big_chances_created', 'key_passes_per_game', 'accurate_per_game', 'acc_own_half', 'acc_opposition_half', 'acc_long_balls', 'acc_chipped_passes', 'acc_crosses', 'clean_sheets', 'interceptions_per_game', 'tackles_per_game', 'possession_won_(final_third)', 'balls_recovered_per_game', 'dribbled_past_per_game', 'clearances_per_game', 'errors_leading_to_shot', 'errors_leading_to_goal', 'penalties_committed', 'succ_dribbles', 'total_duels_won', 'ground_duels_won', 'aerial_duels_won', 'possession_lost', 'fouls', 'was_fouled', 'offsides', 'goal_kicks_per_game', 'yellow', 'yellow_red', 'red_cards', 'average_sofascore_rating'
]
```

## Convert numerical columns to float

```python
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
```

## Filling Missing Values

```python
for col in numerical_columns:
    # Sütunu float'a çevir, hata varsa NaN yap, sonra eksik değerleri 0.0 ile doldur
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(float).fillna(0.0)

# Eksik değerleri kontrol edelim
print(df.isnull().sum())
```

## Checking for the Missing Players

```python
# CSV dosyasını oku
performance_stats_file_path = "../data/raw_data/players_sofascore.csv"
players_file_path = "../data/processed_data/players_raw.csv"

# Verileri yükle
df = pd.read_csv(performance_stats_file_path)
footballers_df = pd.read_csv(players_file_path)

# Mevcut 'footballer_id' değerlerini al
existing_footballer_ids = df['footballer_id'].unique()

# 1'den 1585'e kadar olan 'footballer_id' değerlerini oluştur
all_footballer_ids = set(range(1, 1586))

# Eksik olan 'footballer_id' değerlerini bul
missing_footballer_ids = sorted(all_footballer_ids - set(existing_footballer_ids))

# Eksik 'footballer_id' değerleri ile ilgili veriyi al
missing_players = footballers_df[footballers_df['footballer_id'].isin(missing_footballer_ids)]

# Eksik oyuncuları yeni CSV dosyasına kaydet
missing_players[['footballer_id', 'footballer_name']].to_csv('../data/processed_data/players_sofascore_missing.csv', index=False)

print(f"Toplam eksik 'footballer_id' sayısı: {len(missing_footballer_ids)}")
print("Eksik 'footballer_id' değerleri ve oyuncu bilgileri 'players_missing.csv' dosyasına aktarıldı.")
```
