## Cleaning

```python
# Eksik değer içeren sütunları ve her bir sütundaki eksik değer sayısını göster
missing_data = df.isnull().sum()

# Sadece eksik değeri olan sütunları filtrele
missing_data = missing_data[missing_data > 0]

# Eksik değer içeren sütunlar ve sayıları yazdır
print("Eksik Değer İçeren Sütunlar ve Eksik Değer Sayıları:")
print(missing_data)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Eksik olmayan verileri al
train_data = df.dropna(subset=['potential'])

# Eksik olan verileri al
missing_data = df[df['potential'].isnull()]

# Özellikleri belirle
# Futbolcuların teknik ve fiziksel tüm özelliklerini kullanalım
features = [
    'rating', 'agility', 'stamina', 'goalkeeping', 'mental', 'physical', 
    'technical', 'aerial-reach', 'command-of-area', 'communication', 'eccentricity', 
    'first-touch', 'handling', 'kicking', 'one-on-ones', 'passing', 'punching-tendency', 
    'reflexes', 'rushing-out-tendency', 'throwing', 'aggression', 'anticipation', 'bravery', 
    'composure', 'concentration', 'decisions', 'determination', 'flair', 'leadership', 
    'off-the-ball', 'positioning', 'teamwork', 'vision', 'work-rate', 'acceleration', 'balance', 
    'jumping-reach', 'natural-fitness', 'pace', 'strength', 'free-kick-taking', 'penalty-taking', 
    'technique', 'corners', 'crossing', 'dribbling', 'finishing', 'heading', 'long-shots', 
    'long-throws', 'marking', 'tackling'
]

# Modeli eğit
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(train_data[features], train_data['potential'])

# Eksik verileri tahmin et ve yerine koy
predicted_values = rf.predict(missing_data[features])

# Tahmin edilen değerleri sakla
df.loc[df['potential'].isnull(), 'potential'] = predicted_values

# Tahmin edilen satırları göster
filled_rows = df.loc[missing_data.index, ['potential'] + features]
print("Doldurulan Potential Değerleri:")
print(filled_rows)

# Özelliklerin önem sıralamasını görmek
feature_importance = pd.DataFrame({
    'Feature': features, 
    'Importance': rf.feature_importances_
})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

```
