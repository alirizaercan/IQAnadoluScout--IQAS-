```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for a single player over 30 days
np.random.seed(0)
days = np.arange(1, 31)
vo2_max = 50 + np.random.normal(loc=0, scale=2, size=30)  # VO₂ Max (ml/kg/min)
lactate_levels = 2 + np.random.normal(loc=0, scale=0.2, size=30)  # Lactate levels (mmol/L)

# Line Plot for VO₂ Max
plt.figure(figsize=(10, 5))
plt.plot(days, vo2_max, marker='o', label='VO₂ Max (ml/kg/min)', color='blue')
plt.title('VO₂ Max Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('VO₂ Max (ml/kg/min)', fontsize=14)
plt.xticks(days, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('vo2_max_over_30_days.png', format='png', dpi=300)  # Save as PNG
plt.show()

# Bar Plot for Lactate Levels
plt.figure(figsize=(10, 5))
plt.bar(days, lactate_levels, color='orange')
plt.title('Lactate Levels Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Lactate Levels (mmol/L)', fontsize=14)
plt.xticks(days, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('lactate_levels_over_30_days.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
# Sample data for training intensity and recovery times over 30 days
training_intensity = np.random.randint(60, 100, size=30)  # Training intensity (%)
recovery_times = np.random.randint(24, 48, size=30)  # Recovery times (hours)

# Line Chart for Training Intensity
plt.figure(figsize=(10, 5))
plt.plot(days, training_intensity, marker='o', label='Training Intensity (%)', color='green')
plt.title('Training Intensity Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Training Intensity (%)', fontsize=14)
plt.xticks(days, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('training_intensity_over_30_days.png', format='png', dpi=300)  # Save as PNG
plt.show()

# Scatter Plot for Recovery Times
plt.figure(figsize=(10, 5))
plt.scatter(days, recovery_times, color='purple', s=100)
plt.title('Recovery Times Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Recovery Times (hours)', fontsize=14)
plt.xticks(days, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('recovery_times_over_30_days.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Fit linear regression model for VO₂ Max trend prediction
vo2_max_data = np.array(vo2_max).reshape(-1, 1)
days_reshaped = days.reshape(-1, 1)

model = LinearRegression()
model.fit(days_reshaped, vo2_max_data)
predictions = model.predict(days_reshaped)

# Line Plot with Regression Line
plt.figure(figsize=(10, 5))
plt.plot(days, vo2_max, marker='o', label='VO₂ Max (ml/kg/min)', color='blue')
plt.plot(days, predictions, label='Trend Line', color='red', linestyle='--')
plt.title('VO₂ Max Trend Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('VO₂ Max (ml/kg/min)', fontsize=14)
plt.xticks(days, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('vo2_max_trend_over_30_days.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
# Sample data for progress towards goals
current_values = [55, 2.5, 80]  # Current performance
target_values = [60, 1.8, 90]  # Target performance
goals = ['VO₂ Max', 'Lactate Level', 'Muscle Strength']

# Bar Chart for Progress Towards Goals
x = np.arange(len(goals))
width = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x - width/2, current_values, width, label='Current Values', color='lightblue')
plt.bar(x + width/2, target_values, width, label='Target Values', color='lightgreen')

plt.title('Progress Towards Goals', fontsize=16, fontweight='bold')
plt.xlabel('Goals', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(x, goals, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y')
plt.legend()
plt.tight_layout()
plt.savefig('progress_towards_goals.png', format='png', dpi=300)  # Save as PNG
plt.show()
```
