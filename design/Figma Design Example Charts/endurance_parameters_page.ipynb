```python
# Key Endurance Metrics Overview
# Athletic Profile Summary

import numpy as np
import matplotlib.pyplot as plt

# Sample data for key endurance metrics
days = np.arange(1, 31)
running_distance = np.random.normal(loc=5, scale=1, size=30)  # Average distance in km
average_speed = np.random.normal(loc=10, scale=2, size=30)  # Average speed in km/h
heart_rate = np.random.randint(120, 180, size=30)  # Heart rate in bpm

# Create a figure for metric cards
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Running Distance Card
ax[0].barh(['Player'], [np.mean(running_distance)], color='lightblue')
ax[0].set_xlim(0, 10)
ax[0].set_title('Average Running Distance (km)', fontsize=14)
ax[0].text(np.mean(running_distance) + 0.1, 0, f"{np.mean(running_distance):.2f}", va='center')

# Average Speed Card
ax[1].barh(['Player'], [np.mean(average_speed)], color='lightgreen')
ax[1].set_xlim(0, 15)
ax[1].set_title('Average Speed (km/h)', fontsize=14)
ax[1].text(np.mean(average_speed) + 0.1, 0, f"{np.mean(average_speed):.2f}", va='center')

# Heart Rate Card
ax[2].barh(['Player'], [np.mean(heart_rate)], color='salmon')
ax[2].set_xlim(100, 200)
ax[2].set_title('Average Heart Rate (bpm)', fontsize=14)
ax[2].text(np.mean(heart_rate) + 1, 0, f"{np.mean(heart_rate):.2f}", va='center')

# Adjust layout
plt.tight_layout()
plt.savefig('endurance_metrics_overview.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
# Endurance Trends

# Create a line chart for endurance parameters over 30 days
plt.figure(figsize=(10, 6))
plt.plot(days, running_distance, marker='o', color='royalblue', label='Running Distance (km)')
plt.plot(days, average_speed, marker='o', color='orange', label='Average Speed (km/h)')
plt.plot(days, heart_rate, marker='o', color='red', label='Heart Rate (bpm)')
plt.title('Endurance Metrics Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(days, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('endurance_development_over_time.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
# Peak Heart Rate Focused Endurance Development

import matplotlib.pyplot as plt
import numpy as np

# Sample data
sessions = np.arange(1, 11)  # 10 training sessions
peak_heart_rate = np.random.randint(160, 200, 10)  # Peak heart rate (bpm)

# Create the plot with a professional style (using 'ggplot' style)
plt.style.use('ggplot')  # Using a built-in style from matplotlib

# Create the figure
plt.figure(figsize=(12, 7))

# Plotting the peak heart rate with a bar chart and adding edge color
bars = plt.bar(sessions, peak_heart_rate, color='slateblue', width=0.6, edgecolor='black')

# Adding horizontal lines for clarity
plt.axhline(y=np.mean(peak_heart_rate), color='r', linestyle='--', label=f'Mean: {np.mean(peak_heart_rate):.2f} bpm')

# Title and labels with professional font settings
plt.title('Peak Heart Rate Development Over 10 Sessions', fontsize=18, fontweight='bold')
plt.xlabel('Sessions', fontsize=14, fontweight='bold')
plt.ylabel('Peak Heart Rate (bpm)', fontsize=14, fontweight='bold')

# Add custom ticks and grid
plt.xticks(sessions, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle=':', color='gray', alpha=0.6)

# Adding value labels on top of the bars for clarity
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 2, round(yval, 1), ha='center', va='bottom', fontsize=10)

# Display the plot with a tight layout
plt.legend(loc='upper left', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Performance Radar

import matplotlib.pyplot as plt
import numpy as np

# Gerek veriler
# 30 gnlk kou mesafesi (km), ortalama hz (km/h) ve kalp at hz (bpm)
np.random.seed(0)  # Tekrar edilebilirlik iin rastgelelik
running_distance = np.random.uniform(5, 10, 30)  # 5-10 km arasnda rastgele veriler
average_speed = np.random.uniform(8, 12, 30)  # 8-12 km/s arasnda rastgele veriler
heart_rate = np.random.uniform(130, 150, 30)  # 130-150 bpm arasnda rastgele veriler

# Kategoriler
categories = ['Running Distance (km)', 'Average Speed (km/h)', 'Heart Rate (bpm)']

# Oyuncunun verileri
player_values = [np.mean(running_distance), np.mean(average_speed), np.mean(heart_rate)]

# Dier oyuncularn rnek deerleri (rnek)
other_players_values = [6.5, 10.5, 140]  # rnek deerler

# Deiken says
num_vars = len(categories)

# Alar hesapla
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Verileri emberin kapatlmas iin dngy tamamla
player_values += player_values[:1]
other_players_values += other_players_values[:1]
angles += angles[:1]

# Radar grafii olutur
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, player_values, color='blue', alpha=0.25, label='Player')
ax.fill(angles, other_players_values, color='red', alpha=0.25, label='Other Players')

# Etiketler ekle
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)
ax.set_title('Endurance Comparison (Radar Chart)', fontsize=16, fontweight='bold')
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Arka plan olmadan gster
ax.spines['polar'].set_visible(False)  # Arka plan kaldr
plt.tight_layout()

# PNG olarak kaydet
plt.savefig('endurance_comparison_radar_chart.png', format='png', dpi=300)  # Kaydet
plt.show()
```
