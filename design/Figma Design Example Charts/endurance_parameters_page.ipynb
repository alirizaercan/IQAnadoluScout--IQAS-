```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for key endurance metrics
days = np.arange(1, 31)
running_distance = np.random.normal(loc=5, scale=1, size=30)  # Average distance in km
average_speed = np.random.normal(loc=10, scale=2, size=30)  # Average speed in km/h
heart_rate = np.random.randint(120, 180, size=30)  # Heart rate in bpm

# Create a figure for metric cards
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Running Distance Card
ax[0].barh(['Player'], [np.mean(running_distance)], color='lightblue')
ax[0].set_xlim(0, 10)
ax[0].set_title('Average Running Distance (km)', fontsize=14)
ax[0].text(np.mean(running_distance) + 0.1, 0, f"{np.mean(running_distance):.2f}", va='center')

# Average Speed Card
ax[1].barh(['Player'], [np.mean(average_speed)], color='lightgreen')
ax[1].set_xlim(0, 15)
ax[1].set_title('Average Speed (km/h)', fontsize=14)
ax[1].text(np.mean(average_speed) + 0.1, 0, f"{np.mean(average_speed):.2f}", va='center')

# Heart Rate Card
ax[2].barh(['Player'], [np.mean(heart_rate)], color='salmon')
ax[2].set_xlim(100, 200)
ax[2].set_title('Average Heart Rate (bpm)', fontsize=14)
ax[2].text(np.mean(heart_rate) + 1, 0, f"{np.mean(heart_rate):.2f}", va='center')

# Adjust layout
plt.tight_layout()
plt.savefig('endurance_metrics_overview.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
# Create a line chart for endurance parameters over 30 days
plt.figure(figsize=(10, 6))
plt.plot(days, running_distance, marker='o', color='royalblue', label='Running Distance (km)')
plt.plot(days, average_speed, marker='o', color='orange', label='Average Speed (km/h)')
plt.plot(days, heart_rate, marker='o', color='red', label='Heart Rate (bpm)')
plt.title('Endurance Metrics Over 30 Days', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(days, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('endurance_development_over_time.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample training intensity data (30 sessions, 10 zones)
zones = [f'Zone {i+1}' for i in range(10)]
training_intensity = np.random.randint(0, 10, size=(30, 10))  # Values between 0 and 10 for intensity

# Create a DataFrame for better visualization
data = {
    'Session': np.arange(1, 31),
    'Zone 1': training_intensity[:, 0],
    'Zone 2': training_intensity[:, 1],
    'Zone 3': training_intensity[:, 2],
    'Zone 4': training_intensity[:, 3],
    'Zone 5': training_intensity[:, 4],
    'Zone 6': training_intensity[:, 5],
    'Zone 7': training_intensity[:, 6],
    'Zone 8': training_intensity[:, 7],
    'Zone 9': training_intensity[:, 8],
    'Zone 10': training_intensity[:, 9],
}

df = pd.DataFrame(data)

# Create a heatmap for training intensity
plt.figure(figsize=(12, 6))
heatmap = sns.heatmap(
    df.set_index('Session'), 
    cmap='YlOrRd',  # Using a yellow-red color palette for better visual effect
    annot=True, 
    fmt='d', 
    linewidths=.5, 
    cbar_kws={'label': 'Intensity Level'},
    mask=(df.set_index('Session') == 0)  # Mask to remove annotation for zeros
)

# Title and labels
plt.title('Heatmap of Training Intensity in Training Zones Over 30 Sessions', fontsize=16, fontweight='bold')
plt.xlabel('Training Zones', fontsize=14)
plt.ylabel('Training Sessions', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)  # Keep y-ticks horizontal
plt.tight_layout()

# Save as PNG with better quality
plt.savefig('training_intensity_heatmap_professional.png', format='png', dpi=300)  # Save as PNG
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerçek veriler
# 30 günlük koşu mesafesi (km), ortalama hız (km/h) ve kalp atış hızı (bpm)
np.random.seed(0)  # Tekrar edilebilirlik için rastgelelik
running_distance = np.random.uniform(5, 10, 30)  # 5-10 km arasında rastgele veriler
average_speed = np.random.uniform(8, 12, 30)  # 8-12 km/s arasında rastgele veriler
heart_rate = np.random.uniform(130, 150, 30)  # 130-150 bpm arasında rastgele veriler

# Kategoriler
categories = ['Running Distance (km)', 'Average Speed (km/h)', 'Heart Rate (bpm)']

# Oyuncunun verileri
player_values = [np.mean(running_distance), np.mean(average_speed), np.mean(heart_rate)]

# Diğer oyuncuların örnek değerleri (örnek)
other_players_values = [6.5, 10.5, 140]  # Örnek değerler

# Değişken sayısı
num_vars = len(categories)

# Açıları hesapla
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Verileri çemberin kapatılması için döngüyü tamamla
player_values += player_values[:1]
other_players_values += other_players_values[:1]
angles += angles[:1]

# Radar grafiği oluştur
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, player_values, color='blue', alpha=0.25, label='Player')
ax.fill(angles, other_players_values, color='red', alpha=0.25, label='Other Players')

# Etiketler ekle
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)
ax.set_title('Endurance Comparison (Radar Chart)', fontsize=16, fontweight='bold')
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Arka plan olmadan göster
ax.spines['polar'].set_visible(False)  # Arka planı kaldır
plt.tight_layout()

# PNG olarak kaydet
plt.savefig('endurance_comparison_radar_chart.png', format='png', dpi=300)  # Kaydet
plt.show()
```
