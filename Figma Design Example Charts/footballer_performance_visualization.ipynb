```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from math import pi
```

```python
# Saha ölçülerine göre veri noktaları üret
np.random.seed(42)
x_coords = np.random.normal(50, 10, 500)
y_coords = np.random.normal(25, 8, 500)

# Futbol sahası resmi
field_img = cv2.imread("soccer-field.jpg")
field_img_rgb = cv2.cvtColor(field_img, cv2.COLOR_BGR2RGB)

# Isı haritası grafiği
plt.figure(figsize=(12, 8))
plt.imshow(field_img_rgb, extent=[0, 100, 0, 65])  # Saha sınırları resme göre ayarlanır
sns.kdeplot(x=x_coords, y=y_coords, cmap="Reds", alpha=0.6, fill=True)
plt.title("Valon Ethemi - 90 Minute Heatmap", fontsize=16, color="darkred")
plt.xlabel("Field Width")
plt.ylabel("Field Height")
plt.savefig("ValonEthemi_Heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

# Saha ölçülerine göre veri noktaları üret
np.random.seed(42)

# Sol kanat ve orta saha pozisyonlarına yoğun veri üretimi
x_coords = np.concatenate([
    np.random.normal(175, 20, 300),  # Sol kanat yoğunluğu (320-430 arası)
    np.random.normal(400, 50, 200),   # Orta saha yoğunluğu (300-450 arası)
])
y_coords = np.concatenate([
    np.random.normal(150, 30, 300),   # Sol kanat y ekseni (200-250 arası)
    np.random.normal(250, 40, 200),   # Orta saha y ekseni (240-290 arası)
])

# Futbol sahası resmi
field_img = cv2.imread("soccer-field.jpg")
field_img_rgb = cv2.cvtColor(field_img, cv2.COLOR_BGR2RGB)

# Isı haritası grafiği
plt.figure(figsize=(12, 8))
plt.imshow(field_img_rgb, extent=[0, 751, 0, 490])  # Y ekseninin 0-490 olarak ayarlandığından emin olun

# Isı haritasını oluştur (sadece kırmızı tonları kullanarak)
sns.kdeplot(
    x=x_coords, 
    y=y_coords, 
    cmap="Reds",  # Sadece kırmızı tonları
    alpha=0.7,   # Şeffaflık
    fill=True, 
    levels=60, 
    thresh=0.01
)

# Oyuncu pozisyonu ikonu (örneğin, bir daire ile gösterelim)
plt.scatter(375, 220, color='red', s=200, alpha=0.6)  # Sol kanat
plt.scatter(375, 250, color='darkred', s=200, alpha=0.6)  # Orta saha

# Eksensiz görselleştirme
plt.axis('off')  # Eksenleri gizle

# Görselleştirmeyi kaydet
plt.savefig("ValonEthemi_Heatmap_Red_Professional.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Performans verileri
stats = [
    "Speed", "Passing Accuracy", "Dribbling", "Stamina", "Shooting", 
    "Defensive Actions", "Aerial Duels", "Ball Recoveries", "Tackles Won", 
    "Interceptions", "Clearances", "Cross Accuracy", "Key Passes", 
    "Shots on Target", "Distance Covered", "Possession Time"
]
values = [80, 88, 75, 90, 70, 65, 55, 60, 78, 82, 66, 74, 81, 69, 85, 72]
values += values[:1]  # Grafiği tamamlamak için ilk değer tekrar eklenir

# Açıları hesapla
angles = [n / float(len(stats)) * 2 * pi for n in range(len(stats))]
angles += angles[:1]

plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
ax.set_facecolor("white")

# Radar grafiği çizimi
plt.xticks(angles[:-1], stats, color="darkblue", size=10)
ax.plot(angles, values, color="royalblue", linewidth=2, linestyle='solid')
ax.fill(angles, values, color="lightblue", alpha=0.4)

# Başlık ve dosya kaydetme
plt.title("Valon Ethemi - Detailed Match Performance", fontsize=16, color="darkblue")
plt.savefig("ValonEthemi_DetailedRadarChart.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
periods = ["0-15", "15-30", "30-45", "45-60", "60-75", "75-90"]
distance_covered = [1100, 1400, 1250, 1600, 1350, 1450]

plt.figure(figsize=(10, 6))
plt.plot(periods, distance_covered, color="darkgreen", marker="o", linestyle="-", linewidth=2)
plt.fill_between(periods, distance_covered, color="lightgreen", alpha=0.3)
plt.xlabel("Time Period (minutes)")
plt.ylabel("Distance Covered (meters)")
plt.title("Valon Ethemi - Distance Covered per 15 Minutes", fontsize=14, color="darkgreen")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.savefig("ValonEthemi_DistanceCovered.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
time_periods = ["0-15", "15-30", "30-45", "45-60", "60-75", "75-90"]
possession_times = [5, 8, 7, 6, 7, 5]

plt.figure(figsize=(10, 6))
plt.bar(time_periods, possession_times, color="#FF6347", edgecolor="black")
plt.xlabel("Time Period (minutes)")
plt.ylabel("Possession Time (minutes)")
plt.title("Valon Ethemi - Possession Time per 15 Minutes", fontsize=14, color="darkred")
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.savefig("ValonEthemi_PossessionTime.png", dpi=300, bbox_inches='tight')
plt.show()
```
